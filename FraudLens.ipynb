```python
import pandas as pd
import numpy as np
import torch
```

```python
labels_df=pd.read_csv("elliptic_bitcoin_dataset/elliptic_txs_classes.csv")
edges_df=pd.read_csv("elliptic_bitcoin_dataset/elliptic_txs_edgelist.csv")
features_df=pd.read_csv("elliptic_bitcoin_dataset/elliptic_txs_features.csv",header=None)
```

```python
features_df.rename(columns={0:"txId"},inplace=True)
features_df
```

```python
edges_df
```

```python
labels_df
```

### Features Scaling and txid:idx ratio

```python
from sklearn.preprocessing import StandardScaler
features = features_df.drop(columns=["txId"]).values
scaler = StandardScaler()
features = scaler.fit_transform(features)
features
```

```python
x = torch.tensor(features, dtype=torch.float)
x
```

```python
txid_to_idx={txId:idx for idx,txId in enumerate(features_df["txId"])}
txid_to_idx
```

### edges src->dest in index format

```python
edge_index = torch.tensor([
    [txid_to_idx[src] for src in edges_df["txId1"]],
    [txid_to_idx[dst] for dst in edges_df["txId2"]]
], dtype=torch.long)
edge_index
```

### labels encode and store in dict format for easy access

```python
labels_map={"unknown":-1,"2":0,"1":1}
labels_df["class"]=labels_df["class"].replace(labels_map)
labels_df
```

```python
label_dict=dict(zip(labels_df["txId"],labels_df["class"]))
label_dict
```

```python
labels = [label_dict.get(tx_id, -1) for tx_id in features_df["txId"]]
labels
```

```python
labels=torch.tensor(labels,dtype=torch.long)
labels
```

## Create pytorch geometric data obj

```python
from torch_geometric.data import Data
data= Data(x=x, edge_index=edge_index,y=labels)
data
```

## Split

```python
from sklearn.model_selection import train_test_split

labeled_idx=(data.y != -1).nonzero(as_tuple=True)[0]
y_labeled=data.y[labeled_idx]
y_labeled
```

```python
# Split into train/val/test
train_idx, temp_idx = train_test_split(labeled_idx, test_size=0.3, stratify=y_labeled, random_state=42)
val_idx, test_idx = train_test_split(temp_idx, test_size=0.5, stratify=data.y[temp_idx], random_state=42)
```

```python
# create masks
#initally all nodes are false
data.train_mask=torch.zeros(data.num_nodes,dtype=torch.bool)
data.val_mask=torch.zeros(data.num_nodes,dtype=torch.bool)
data.test_mask=torch.zeros(data.num_nodes,dtype=torch.bool)

#make true for some nodes
data.train_mask[train_idx] = True
data.val_mask[val_idx] = True
data.test_mask[test_idx] = True

print(f"Train: {data.train_mask.sum().item()}, Val: {data.val_mask.sum().item()}, Test: {data.test_mask.sum().item()}")
```

## Define GNN model 1)define layers 2)move to device 3)define fwd funcn

```python
from torch_geometric.nn import GCNConv

conv1=GCNConv(in_channels=166,out_channels=64)
conv2=GCNConv(in_channels=64,out_channels=2)

device= torch.device('cuda' if torch.cuda.is_available() else 'cpu')
conv1=conv1.to(device)
conv2=conv2.to(device)
data=data.to(device)
```

```python
import torch.nn.functional as AF

def forward(data):
    x, edge_index=data.x,data.edge_index
    x=conv1(x,edge_index)
    x=AF.relu(x)
    x=conv2(x,edge_index)
    return x
```

## Model Training

```python
import torch.optim as optim
import torch.nn as nn

optimizer=optim.Adam(list(conv1.parameters())+list(conv2.parameters()),lr=0.01,weight_decay=5e-4)
criterion=nn.CrossEntropyLoss()
```

```python
def train():
    conv1.train()
    conv2.train()
    optimizer.zero_grad()
    out=forward(data)
    loss=criterion(out[data.train_mask],data.y[data.train_mask])
    loss.backward()
    optimizer.step()
    return loss.item()
```

```python
def evaluate(mask):
    conv1.eval()
    conv2.eval()
    out=forward(data)
    preds=out[mask].argmax(dim=1)
    acc=(preds == data.y[mask]).sum().item() / mask.sum().item()
    return acc
```

```python
for epoch in range(1, 201):
    loss = train()
    val_acc = evaluate(data.val_mask)
    if epoch % 20 == 0:
        print(f"Epoch {epoch:03d}, Loss: {loss:.4f}, Val Acc: {val_acc:.4f}")
```

```python
test_acc = evaluate(data.test_mask)
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
from sklearn.metrics import classification_report

conv1.eval()
conv2.eval()
out = forward(data)

preds = out[data.test_mask].argmax(dim=1).cpu().numpy()
true = data.y[data.test_mask].cpu().numpy()

print(classification_report(true, preds, target_names=["licit", "illicit"]))
```
